```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0, 10, 11)
b = a**4
```

```python
x = np.arange(0,10)
y = 2 *x
```

```python
fig, axes = plt.subplots(nrows = 1, ncols=2)
axes[0].plot(x, y)
axes[1].plot(a, b)
```

```python
fig, axes = plt.subplots(nrows= 2, ncols = 2)
axes[0][0].plot(x, y)
axes[0][1].plot(a, b)
axes[1][0].plot(x, y)
axes[1][1].plot(a, b)

fig.suptitle("Figure Level", fontsize = 16)

fig.subplots_adjust(wspace=0.4, hspace=0.4)
fig.savefig('new_subplot.png', bbox_inches = 'tight')
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 1)
for ax in axes:
    ax.plot(x, y)
    
plt.tight_layout()
```
