```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 11, 10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x, label= 'X vs X')
ax.plot(x, x**2, label = 'X vs X^2')
ax.legend(loc=0)
```

```python
fig = plt.figure()
ax =fig.add_axes([0,0,1,1])
ax.plot(x,x, color = 'red', label= 'X  vs X')
ax.plot(x, x**2, color = '#1313c2', label = 'X vs X^2') #Hex Color
ax.legend()
```

```python
fig = plt.figure()
ax =fig.add_axes([0,0,1,1])
ax.plot(x,x, color = 'red', label= 'X  vs X', lw = 10, ls = '--')
ax.plot(x, x + 1, color = '#1313c2', label = 'X vs X^2', linewidth = 5, linestyle = '-.') #Hex Color
ax.legend()
```

```python
fig = plt.figure()
ax =fig.add_axes([0,0,1,1])
ax.plot(x, x + 1, marker = 'o', ls = '--') 
ax.plot(x, x, marker = 'x', markersize = 10,  ls = '--') 
```

```python
fig = plt.figure()
ax =fig.add_axes([0,0,1,1])
ax.plot(x, x + 1, marker = 's', ms =15 ,ls = '--', markerfacecolor = 'red', markeredgewidth = 4, markeredgecolor = 'orange') 
```
